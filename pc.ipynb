```python
import mysql


import mysql.connector
import requests
from bs4 import BeautifulSoup
import pandas as pd
names=[]
prices=[]
image=[]



for i in  range(1,10):
    page = requests.get("https://www.jumia.ma/smartphones-android/?page="+str(i))
    soup = BeautifulSoup(page.content, 'html.parser')
    items = soup.find_all('div', attrs={'class': 'info'})
    items2 = soup.find_all('div', attrs={'class': 'img-c'})
    names=names + [item.find(class_='name').get_text() for item in items]
    prices=prices + [item.find(class_='prc').get_text() for item in items]
    image=image + [item.find(class_='img').get('data-src')   for item in items2]
    print('Web scarping ......... page:',i )




```

```python
db=mysql.connector.connect(user="root",database="scraping_sample")
cursor = db.cursor()

add_news=("Insert into jumia(name,price,image) values(%s,%s,%s)")

data_news =(names,prices,image)
print(data_news)




#df=pd.DataFrame({
   # 'nom de telephone':names,
    #'prix':prices,
    #'image':image,
#} )
#df.to_csv('telephones.csv')
```

```python
cursor.execute(add_news,data_news).join(list)




```
