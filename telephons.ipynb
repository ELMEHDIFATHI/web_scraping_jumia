```python
import  requests
from  bs4 import BeautifulSoup
import  pandas as pd
response = page=requests.get('https://www.jumia.ma/catalog/?q=telephone+')
soup=BeautifulSoup(page.content,'html.parser')
rowOfArticl = soup.find(class_='-paxs row _no-g _4cl-3cm-shs')
print(rowOfArticl.prettify())
```

```python
articles = rowOfArticl.find_all(class_='prd _fb col c-prd')
print(articles)
```

```python
article_names =[article.find(class_='name').get_text() for article in articles]
print(article_names)

article_prices =[article.find(class_='prc').get_text() for article in articles]
print(article_prices)

article_images= [article.find(class_='img').get('data-src')   for article in articles]
print(article_prices)
```

```python
articls_df = pd.DataFrame(
    {'name': article_names,
     'price': article_prices,
     'image':article_images}
)
print(articls_df)
articls_df.to_csv('jumia_first_page.csv')
```

```python
productNames=[]
productPrice=[]
productImage=[]
for i in range(1,15):
    response = page=requests.get('https://www.jumia.ma/catalog/?q=telephone+&page='+str(i)+'#catalog-listing')
    soup=BeautifulSoup(page.content,'html.parser')
    allarticle = soup.find(class_='-paxs row _no-g _4cl-3cm-shs')
    articles = allarticle.find_all(class_='prd _fb col c-prd')
    for article in articles:
        productNames.append(article.find(class_='name').get_text())
        productPrice.append(article.find(class_='prc').get_text())
        productImage.append(article.find(class_='img').get('data-src'))

articls_df = pd.DataFrame(
    {'name': productNames,
     'price': productPrice,
     'image':productImage,

     }
)
print(articls_df)
articls_df.to_csv('jumia_phone.csv')

```
